```
import numpy as np
from jupyterthemes.jtplot import figsize
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```
data = fetch_olivetti_faces(shuffle=True, random_state=69)
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=69)
model = GaussianNB()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

fig, axes = plt.subplots(3, 5, figsize=(6, 6))
for ax, img, lbl, pred in zip(axes.ravel(), x_test, y_test, y_pred):
    ax.imshow(img.reshape(64, 64), cmap="gray")
    ax.axis("off")
    ax.set_title(f"T: {lbl}, P: {pred}")
plt.show()

```
