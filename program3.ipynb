```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```
iris = load_iris()
data = iris.data
labels = iris.target
label_names = iris.target_names
iris_df = pd.DataFrame(data, columns=iris.feature_names)
```

```
pca = PCA(n_components=2)
data_reduced = pca.fit_transform(data)
reduced_df = pd.DataFrame(data_reduced, columns=["principal_component_1", "principal_component_2"])
reduced_df["label"] = labels
```

```
from bokeh.colors.named import colors

# plot the reduced data
plt.figure(figsize=(15, 10))
for idx, label in enumerate(np.unique(labels)):
    plt.scatter(
        reduced_df[reduced_df['label'] == label]['principal_component_1'],
        reduced_df[reduced_df["label"] == label]["principal_component_2"],
        label=label_names[label],
    )

plt.grid()
plt.legend()
plt.title("iris Dataset")
plt.show()
```
